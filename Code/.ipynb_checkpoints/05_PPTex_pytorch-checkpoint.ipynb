```python
import torch
import torch.nn as nn
import torch.optim as optimizer
import numpy as np
import matplotlib.pyplot as plt
```

```python
#) data
event_id = []
event_feature = []
event_target = []
with open ("ECS_pytorch.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()
    for line in lines:
        event_id.append(line.split()[0])
        event_feature.append([float(line.split()[1])])
        event_target.append([float(line.split()[2])])
```

```python
import torch.nn as nn

class LinearRegression(nn.Module):
    def __init__(self, input_dem, output_dem):
        super().__init__()

        self.linear = nn.Linear(input_dem, output_dem)
        
    def forward(self, x):
        return self.linear(x)

#) pre-processing
X = torch.tensor(event_feature, dtype=torch.float32).reshape(-1, 1)
Y = torch.tensor(event_target, dtype=torch.float32).reshape(-1, 1)

X_log = torch.log(X + 1e-6)
Y_log = torch.log(Y + 1e-6)

X_mean, X_std = X_log.mean(), X_log.std()
Y_mean, Y_std = Y_log.mean(), Y_log.std()

X_norm = (X_log - X_mean) / X_std
Y_norm = (Y_log - Y_mean) / Y_std

feature = X_norm
target = Y_norm
```

```python
n_feature = 1
model = LinearRegression(n_feature, 1)

#) loss and optimizer
learning_rate = 0.01
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

#) training
epochs = 1000
for epoch in range(epochs):
    y_predicted = model(feature)
    loss = criterion(y_predicted, target)
    
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if (epoch + 1) % 100 == 0:
        print(f'epoch: {epoch+1}, loss = {loss.item(): .4f}')
```

```python
y_pred = model(feature)
y_pred_original = torch.exp(y_pred * Y_std + Y_mean)  # 反標準化 + 還原 log

with torch.no_grad():
    y_pred = model(feature)
    y_pred_orig = torch.exp(y_pred * Y_std + Y_mean)
    Y_orig = Y
    
    plt.figure(figsize=(6,4))
    plt.scatter(X.numpy(), Y_orig.numpy(), label='Observed', alpha=0.6)
    plt.scatter(X.numpy(), y_pred_orig.numpy(), label='Predicted')
    plt.plot([0, X.max()], [0, X.max()], linestyle='--', color='gray', label='y = x')
    plt.xlabel("Observed PGA")
    plt.ylabel("Predicted PGA")
    plt.legend()
    plt.title("ECS Prediction vs Observation")
    plt.grid(True)
    plt.savefig("ECS_pytorch.png", dpi = 300)
    plt.show()

w_norm = model.linear.weight.item()
b_norm = model.linear.bias.item()
w_orig = w_norm * (Y_std / X_std)
b_orig = b_norm * Y_std + Y_mean - w_orig * X_mean
print(f"訓練完成！\n標準化回歸式：ŷ = {w_norm:.4f} * x_norm + {b_norm:.4f}")
print(f"還原回原始資料的回歸式：y = {w_orig:.4f} * x + {b_orig:.4f}")
```
