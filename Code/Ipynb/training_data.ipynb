```python
from obspy import read, UTCDateTime, Stream
```

```python
st = read("/home/sysop/pytorch/Data/0130Dapu.mseed")
station = "SGS"
t1 = UTCDateTime(2025, 1, 30, 2, 11, 54)
dt1 = 10
dt2 = 15

st2 = st.slice(t1+dt1, t1+dt2)
st3 = st2.normalize()
st4 = st3.detrend("linear")
st5 = st4.detrend("demean")

for i in st5:
    if i.stats.station == station:
        i.plot();
```

```python
st3 = st.normalize()
st4 = st3.detrend("linear")
st5 = st4.detrend("demean")
st6 = Stream()
for tr in st5:
    if tr.stats.npts >= 6000:
        st6.append(tr)
st7 = st6.slice(t1 + dt1, t1 + dt2)

for i in st7:
    station = i.stats.station
    channel = i.stats.channel
    i.plot(outfile = "/home/sysop/pytorch/Output/0130Dapu/5s/" + station + "_" + channel + ".png")  
    i.write(filename = "/home/sysop/pytorch/Output/0130Dapu/5s_mseed/" + station + , format="MSEED")
    print(station, channel)
```

```python
st3 = st.normalize()
st4 = st3.detrend("linear")
st5 = st4.detrend("demean")
st6 = Stream()
for tr in st5:
    if tr.stats.npts >= 6000:
        st6.append(tr)

import matplotlib.pyplot as plt
import numpy as np

for i in st6: 
    fig, ax = plt.subplots(figsize = (10, 3))
    station = i.stats.station
    channel = i.stats.channel
    num_ticks = 5
    tick_times = np.linspace(i.stats.starttime.timestamp, i.stats.endtime.timestamp, num_ticks)
    rounded_tick_times = np.round(tick_times)
    tick_labels = [UTCDateTime(t).strftime('%H:%M:%S') for t in rounded_tick_times]
    
    sample_rate = i.stats.sampling_rate
    sample_index = dt1 * sample_rate
    sample_indexs = dt2 * sample_rate
    
    ax.axvspan(sample_index, sample_indexs, 
               facecolor='lightgreen', # 
               alpha=0.7,              #  70%
               label='Highlighted Interval')
    
    ax.plot(np.linspace(0, i.stats.npts, 6000), i.data)
    ax.set_title(f"Station: {station}, Channel: {channel}")
    ax.set_xlabel("Time (UTC)")
    ax.set_ylabel("Normalized Amplitude")
    ax.set_xticks(np.linspace(0, i.stats.npts, num_ticks))
    ax.set_xticklabels(tick_labels, rotation=30)
    plt.tight_layout()
    plt.savefig("/home/sysop/pytorch/Output/0130Dapu/Full/" + station + "_" + channel + ".png", dpi=300)
    plt.close(fig)
    print(station, channel)
```
